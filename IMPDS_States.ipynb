```python
import pandas as pd, requests, time, os
from io import StringIO
from google.colab import files

# ===== SELECT STATE (Uncomment ONE only) =====
# STATE_NAME, STATE_CODE = "Madhya Pradesh", 23
STATE_NAME, STATE_CODE = "Uttar Pradesh", "09"
# STATE_NAME, STATE_CODE = "Maharashtra", 27
# STATE_NAME, STATE_CODE = "Bihar", 10
# STATE_NAME, STATE_CODE = "Rajasthan", "08"
# STATE_NAME, STATE_CODE = "Gujarat", 24
# STATE_NAME, STATE_CODE = "Karnataka", 29
# STATE_NAME, STATE_CODE = "Tamil Nadu", 33
# STATE_NAME, STATE_CODE = "West Bengal", 19
# STATE_NAME, STATE_CODE = "Andhra Pradesh", 28
# STATE_NAME, STATE_CODE = "Telangana", 36
# STATE_NAME, STATE_CODE = "Kerala", 32
# STATE_NAME, STATE_CODE = "Odisha", 21
# STATE_NAME, STATE_CODE = "Chhattisgarh", 22
# STATE_NAME, STATE_CODE = "Jharkhand", 20
# STATE_NAME, STATE_CODE = "Haryana", "06"
# STATE_NAME, STATE_CODE = "Punjab", "03"
# STATE_NAME, STATE_CODE = "Uttarakhand", "05"
# STATE_NAME, STATE_CODE = "Himachal Pradesh", "02"
# STATE_NAME, STATE_CODE = "Jammu and Kashmir", "01"
# STATE_NAME, STATE_CODE = "Delhi", "07"
# STATE_NAME, STATE_CODE = "Assam", 18
# STATE_NAME, STATE_CODE = "Goa", 30
# STATE_NAME, STATE_CODE = "Meghalaya", 17
# STATE_NAME, STATE_CODE = "Tripura", 16
# STATE_NAME, STATE_CODE = "Mizoram", 15
# STATE_NAME, STATE_CODE = "Nagaland", 13
# STATE_NAME, STATE_CODE = "Arunachal Pradesh", 12
# STATE_NAME, STATE_CODE = "Sikkim", 11
# STATE_NAME, STATE_CODE = "Ladakh", 37
# STATE_NAME, STATE_CODE = "Lakhshadweep", 31
# STATE_NAME, STATE_CODE = "Daman & Diu", 38

# ============================================

# Time period
start_year, end_year = 2020, 2025
print(f"üìä Extracting {STATE_NAME} data...")

# Extract data
data = {k: [] for k in ['T1', 'T2', 'T3', 'T4']}
for year in range(start_year, end_year + 1):
    for month in range(1, 13):
        if year == 2025 and month > 11: continue
        try:
            s = requests.Session()
            s.headers = {"User-Agent": "Mozilla/5.0"}
            s.get(f"https://impds.nic.in/sale/stateUnautmated?month={month}&year={year}")
            r = s.get(f"https://impds.nic.in/sale/stateByCountryAjax?stateCode={STATE_CODE}")
            tables = pd.read_html(StringIO(r.text))
            for i, df in enumerate(tables[:4]):
                df = df.copy()
                df['Month'], df['Year'], df['State'] = month, year, STATE_NAME
                data[list(data.keys())[i]].append(df)
            print(f"‚úÖ {month:02d}/{year}")
        except: print(f"‚ùå {month:02d}/{year}")
        time.sleep(1)

# Download files
base = STATE_NAME.replace(" ", "_")
files_list = [
    (f"{base}_Card_Beneficiary_Stats_2020_2025.csv", data['T1']),
    (f"{base}_Transaction_Counts_2020_2025.csv", data['T2']),
    (f"{base}_Transacted_Ration_Cards_2020_2025.csv", data['T3']),
    (f"{base}_Distributed_Quantity_2020_2025.csv", data['T4'])
]

print("\n‚¨áÔ∏è  Downloading files...")
for filename, data_list in files_list:
    if data_list:
        df = pd.concat(data_list, ignore_index=True)
        df.to_csv(filename, index=False)
        files.download(filename)
        print(f"‚úì {filename} ({len(df)} rows)")

print(f"\n‚úÖ {STATE_NAME} data extraction complete!")
```
